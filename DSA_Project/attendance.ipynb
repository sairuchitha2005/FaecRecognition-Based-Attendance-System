```python
import face_recognition
import os
```

```python
print(os.listdir("/content/drive/MyDrive/ML Dataset 2"))
```

```python
face_enc_dict = {}
face_enc_list = []
for img in os.listdir('/content/drive/MyDrive/ML Dataset 2'):
  path = os.path.join("/content/drive/MyDrive/ML Dataset 2",img)
  face_enc = face_recognition.face_encodings(face_recognition.load_image_file(path))
  face_enc_dict[img[:9]]=face_enc
  face_enc_list.append(face_enc)
  print(f"Done {img[:-4]}")
```

```python
print(face_enc_dict)
```

```python
import pickle

with open("/content/drive/MyDrive/pickle.pkl","wb") as pickle_file:
    pickle.dump(face_enc_dict,pickle_file)
```
